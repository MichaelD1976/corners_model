```python
# Create column 'Proj_HC' and 'Proj_AC' which is the average of the previous 1 calendar year for each division
# So Prem matches in 2020 will have the average HC and AC for matches played 1/1/19 to 31/12/19

from datetime import datetime, timedelta

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

# Function to calculate historical average HC
def calculate_avg_hc(df, div, start_date, end_date):
    period_mask = (df['Div'] == div) & (df['Date'] >= start_date) & (df['Date'] <= end_date)
    avg_value_h = df.loc[period_mask, 'HC'].mean()
    return avg_value_h 

# Function to calculate historical average AC
def calculate_avg_ac(df, div, start_date, end_date):
    period_mask = (df['Div'] == div) & (df['Date'] >= start_date) & (df['Date'] <= end_date)
    avg_value_a = df.loc[period_mask, 'AC'].mean()
    return avg_value_a 

# Function to add 'Proj_HC' and 'Proj_AC' column based on historical averages
def add_proj_hc_ac(df):
    df['Proj_HC'] = None  # Initialize the column with None
    df['Proj_AC'] = None
    
    # Loop through each row and calculate historical averages
    for index, row in df.iterrows():
        div = row['Div']
        current_date = row['Date']
        prev_two_years_start_date = current_date.replace(year=current_date.year - 1, month=1, day=1)
        prev_two_years_end_date = current_date.replace(year=current_date.year - 1, month=12, day=31)
        
        # Calculate the historical average HC for the previous calendar year
        historical_avg_hc = calculate_avg_hc(df, div, prev_two_years_start_date, prev_two_years_end_date)
 
        # Calculate the historical average AC for the previous calendar year
        historical_avg_ac = calculate_avg_ac(df, div, prev_two_years_start_date, prev_two_years_end_date)
        
        # Assign the calculated averages 
        df.at[index, 'Proj_HC'] = round(historical_avg_hc,2)
        df.at[index, 'Proj_AC'] = round(historical_avg_ac,2)
    
    return df

# Assign as df
df_with_proj = add_proj_hc_ac(df)
```
